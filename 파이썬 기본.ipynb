# 5. 리스트 연산 심화

```python
a=[1,2,3,4,5]
b=[3,4,5,6,7]
print(a,b,sep='\n')
```

```python
c=[x for x in a]
print(c)
```

```python
d=[y+1 for y in b]
print(d)
```

```python
e=[x for x in a if x in b]
print(e)
```

```python
f=[y for y in a if y>3]
print(f)
```

```python
#QUIZ!
# b에 있는 값들 중에서 a에는 없는 값
```

```python
g=[x for x in b if x not in a]
print(g)
```
