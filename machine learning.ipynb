```python
# STEP 01 Import libraries
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier # 결정 트리 기반의 앙상블 모델인 랜덤포레스트 회귀/분류 모델입니다
from xgboost import XGBRegressor, XGBClassifier # 부스팅 기반의 xgboost 회귀/분류 모델입니다

# STEP 02 Load dataset
dataset = datasets.load_breast_cancer()

# STEP 03 Split into X, y
X, y = dataset['data'], dataset['target']

# STEP 04 Scaling
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# STEP 05 Split dataset X, y into train, test set
X_train, X_test, y_train, y_test = train_test_split(X_std, y, stratify=y, random_state=42)

# STEP 06 Modeling
model_01 = RandomForestClassifier()
model_02 = XGBClassifier()
model_03 = RandomForestRegressor()
model_04 = XGBRegressor()

# STEP 07 Training
model_01.fit(X_train, y_train)
model_02.fit(X_train, y_train)
model_03.fit(X_train, y_train)
model_04.fit(X_train, y_train)

# STEP 08 Validation
print("학습 스코어:")
print("[분류]")
print(model_01.score(X_train, y_train))
print(model_02.score(X_train, y_train))
print("[회귀]")
print(model_03.score(X_train, y_train))
print(model_04.score(X_train, y_train))

print("\n테스트 스코어:")
print("[분류]")
print(model_01.score(X_test, y_test))
print(model_02.score(X_test, y_test))
print("[회귀]")
print(model_03.score(X_test, y_test))
print(model_04.score(X_test, y_test))
```

## step 1.EDA&Scaling

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

dataset= datasets.load_breast_cancer()

X, y= dataset['data'],dataset['target']

scaler=StandardScaler()
X_std=scaler.fit_transform(X)

X_train,X_test,y_train,y_test=train_test_split(X_std,y,stratify=y,random_state=42)
```

## step 2.Modeling

```python
# STEP 06 Modeling
model_01 = RandomForestClassifier()
model_02 = XGBClassifier()
model_03 = RandomForestRegressor()
model_04 = XGBRegressor()

# 초매개변수 및 default를 확인합니다
print("랜덤포레스트 분류모델:", model_01.get_params())
print("XGBoost 분류모델:", model_02.get_params())
print("랜덤포레스트 회귀모델:", model_03.get_params())
print("XGBoost 회귀모델 :", model_04.get_params())
```

- 이때 주의해야할 점은 ***RandomForest* 모델의 경우, 위에서 설명한 바와 같이 트리 기반의 여러 모델들을 배깅 방식으로 앙상블해 사용하는 알고리즘**이며 이 때문에 학습을 진행할 때마다 모델의 성능이 바뀌게 됩니다
- 이로 인해 ***일반화 성능이 높아지나,*** ***모델 성능 재현의 어려움과 같은 임의성이 존재***합니다.
- ***XGBoost***의 경우, 학습이 반복될 때마다 이전 과정에서 잘 맞추지 못한 부분을 보다 집중적으로 학습하는 **Boosting** 알고리즘으로써 ***학습을 반복할수록 성능이 높아진다***는 장점이 있지만, ***지나치게 반복되면 성능이 올라가는 것처럼 보여도 과대적합될 가능성이 높아져 일반화 가능성이 떨어지게 됩니다***
- 여기서 **과대적합**이란 학습데이터에 해당 알고리즘이 지나치게 적합되어, 학습 데이터 외의 패턴 등에 대한 예측력이 떨어지는 것을 말하며, 학습 데이터를 기반으로 대략적인 패턴을 학습해 예측력을 높게 유지하는 정도를 일반화 성능이라 합니다

```python
# 로지스틱선형회귀모델입니다.

from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression

lr= LogisticRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

print(f"해당 모델의 Roc_auc_score : {roc_auc_score(y_test, y_pred):.4f}")
```

하이퍼파라미터 튜닝

```python
lr.get_params()
```

solver : [‘newton-cg’, ‘lbfgs’, ‘liblinear’, ‘sag’, ‘saga’] # default= 'lbfgs'

penalty : [‘l1’, ‘l2’, ‘elasticnet’, ‘none’] # default= 'l2' - penalty는 어떤 solver를 사용하느냐에 따라 option이 제한됩니다.

C : positive float, default=1.0 - 정규화 강도를 의미하며, 값이 낮을수록 더 강한 정규화를 적용합니다.

l1_ratio : 0 <= l1_ratio <= 1, float # default = None - 'penalty = elasticnet'일 때만 사용가능


```python
from sklearn.metrics import mean_squared_error
```

```python
solver = ['newton-cg','lbfgs','sag']
penalty = ['l2','none']
C = [0.1, 0.5, 1]

for s in solver:
  for p in penalty:
    for c in C:
      print(f"solver = {s}, penalty = {p}, C = {c}일 때의 결과입니다.")
      lr = LogisticRegression(solver = s, penalty=p, C=c)
      lr.fit(X_train, y_train)

      y_pred=lr.predict(X_test)
      
      print("훈련 데이터 점수 :", lr.score(X_train, y_train)) 
      print("테스트 데이터 점수 :", lr.score(X_test,y_test))
      print("roc-auc score : ", roc_auc_score(y_test, y_pred))
      print("="*80)
      
```

```python

solver = ['liblinear'] # 같은 penalty 옵션을 가진 solver들입니다.
penalty = ['l1', 'l2']
C = [0.1, 0.5, 1]

for s in solver :    
  for p in penalty :
    for c in C :
      print(f"solver = {s}, penalty = {p}, C={c}일 때의 결과입니다.")
      lr = LogisticRegression(solver=s,penalty=p, C=c) 
      lr.fit(X_train, y_train) 

      y_pred = lr.predict(X_test)
      
      print("훈련 데이터 점수 :", lr.score(X_train, y_train)) 
      print("테스트 데이터 점수 :", lr.score(X_test,y_test))
      print("roc-auc score : ", roc_auc_score(y_test, y_pred))
      print("="*80)
```

```python
solver = ['saga'] # 같은 penalty 옵션을 가진 solver들입니다.
penalty = ['l1', 'l2', 'elasticnet', 'none']
C = [0.1, 0.5, 1]
l1s = [0, 0.5, 1]

for s in solver :    
  for p in penalty :
    for c in C :
      for l in l1s :          
        print(f"solver = {s}, penalty = {p}, C={c}, l1_ratio = {l}일 때의 결과입니다.")
        lr = LogisticRegression(solver=s,penalty=p, C=c, l1_ratio=l) 
        lr.fit(X_train, y_train) 

        y_pred = lr.predict(X_test)
        
        print("훈련 데이터 점수 :", lr.score(X_train, y_train)) 
        print("테스트 데이터 점수 :", lr.score(X_test,y_test))
        print("roc-auc score : ", roc_auc_score(y_test, y_pred))
        print("="*80)
```
