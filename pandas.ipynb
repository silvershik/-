```python
'''
sample_data의 california_housing_train 데이터를 불러와
housing_median_age를 이용해 십년 단위로 연차를 구분해 
'housing_period'라는 이름의 칼럼에 저장하세요
단위는 '+ years' 를 뒤에 붙이고, 숫자는 소수점이 없는 정수형으로 나타내세요.
이때 위에서 배운 세 가지 방법을 모두 사용해서 풀어보세요.
ex) 15.0 -> 10+ years, 20.0 -> 20+ years
'''
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/sample_data/california_housing_train.csv")
df['housing_period']=(df['housing_median_age']//10*10)
df['housing_period']=df['housing_period'].map(lambda x:str(int(x))+ '+ years')
df
```

```python
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df['housing_period'] = (df['housing_median_age'] // 10 * 10).astype('int').astype('str')
df['housing_period'] = df['housing_period']  + '+ years'
print(df)
```

```python

def modifier(x) :
  return str(int(x) // 10 * 10) + "+ years"

df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df['housing_period'] = df['housing_median_age'].apply(modifier)
print(df)
```

## 예제 2


```python
'''
위에서 생성한 'housing_period' 칼럼을 이용해 20년 미만의 경우 good, 
20년 이상 40년 미만의 경우 normal, 40년 이상의 경우 bad
라는 결과를 가진 'period_grade' 칼럼을 생성하세요.
(사용자 정의 함수를 사용하는 것이 쉬우며, 
map과 lambda를 사용하는 경우에는 lambda 내부의 
조건문 사용이 필요합니다.)
'''
```

```python
def choose(x):
  x = int(x[:-7])

  if x < 20 :
    return "good"
  elif x < 40 :
    return "normal"
  else :
    return "bad"

df['period_grade'] = df['housing_period'].apply(choose)
print(df)
```

## 예제 3

```python
'''
period_grade가 good인 row의 'house_price' 가격의 평균과 
bad인 row의 'house_price' 가격의 평균을 각각 구하고, 그 차이를 구하세요
'''
```

```python
good=df[df['period_grade']=='good']['median_house_value'].mean().round()
bad=df[df['period_grade']=='bad']['median_house_value'].mean().round()
diff=good-bad
print(good, bad, diff)
```
